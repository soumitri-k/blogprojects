```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
import random

random.seed(1234)
```

```python
import pandas as pd

df = pd.read_csv("C:\\Users\\soumi\\Downloads\\housing\\housing.csv")
```

```python
# One-hot encode the string columns
df = pd.get_dummies(df, columns=["ocean_proximity"])
```

```python
# Assume you have a DataFrame 'df' with features and a column 'target' containing class labels
X = df.drop(columns=['median_house_value'])
y = df['median_house_value']
```

```python
X
```

```python
y.shape
```

```python
# Split the data while maintaining class proportions
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Assume you have X_train, X_test, y_train, y_test as your data
regr = LinearRegression()
regr.fit(X_train, y_train)

y_pred = regr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)
print(f"R-squared score: {score:.4f}")
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse:.4f}")
```
